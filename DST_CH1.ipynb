#### Exercise 1.2

Given the following standard form:

\begin{flalign}
\qquad x(t) &= A \cdot \cos( 2\pi \cdot f \cdot t + \phi)  &\text{(1.3)}&
\end{flalign}

\begin{flalign}
\text{1.} \quad x(t) &= \sin( 2\pi \cdot 3 \cdot t) \qquad \rightarrow \qquad x(t) = cos(2\pi \cdot 3 \cdot t - \frac{\pi}{2})&
\end{flalign}

```python
t = np.arange(0, 1.0+1e-5, 1e-5)

x1 = np.sin(2 * np.pi * 3. * t)
x2 = np.cos(2 * np.pi * 3. * t - np.pi/2.)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))


axes[0].plot(t, x1, color='#67a9cf')
axes[0].axhline(y=0, color='grey', alpha=0.1)
axes[0].axvline(x=1, color='grey', alpha=0.1, linestyle='--')
axes[0].set_xlim(-0.005, 1.005)
axes[0].set_ylim(bottom=-1.05, top=1.05)

axes[1].plot(t, x2, color='#ef8a62')
axes[1].get_yaxis().set_visible(False)
axes[1].axhline(y=0, color='grey', alpha=0.1)
axes[1].axvline(x=1, color='grey', alpha=0.1, linestyle='--')
axes[1].set_xlim(-0.005, 1.005)
axes[1].set_ylim(bottom=-1.05, top=1.05)

plt.show()
```

where 

\begin{flalign}
\qquad A &= 1 & \\
\qquad f &= 3 \text{ Hz} &\\
\qquad \phi &= -\frac{\pi}{2} \text{ radians} & 
\end{flalign}

<hr style="width:50%;border-top:1px dashed;"/>

\begin{flalign}
\text{2.} \quad x(t) &= 2 \cdot \cos( \pi \cdot t - \frac{\pi}{2}) \qquad \rightarrow \qquad x(t) = 2 \cdot cos(2\pi \cdot \frac{t}{2} - \frac{\pi}{2})&
\end{flalign}

```python
t = np.arange(0, 2.0+1e-5, 1e-5)

x1 = 2 * np.cos(np.pi * t - np.pi/2.)
x2 = 2 * np.cos(2*np.pi * t/2. - np.pi/2.)

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))


axes[0].plot(t, x1, color='#67a9cf')
axes[0].axhline(y=0, color='grey', alpha=0.1)
axes[0].axvline(x=1, color='grey', alpha=0.1, linestyle='--')
axes[0].axvline(x=2, color='grey', alpha=0.1, linestyle='--')
axes[0].set_xlim(-0.005, 2.005)
axes[0].set_ylim(bottom=-2.05, top=2.05)

axes[1].plot(t, x2, color='#ef8a62')
axes[1].get_yaxis().set_visible(False)
axes[1].axhline(y=0, color='grey', alpha=0.1)
axes[1].axvline(x=1, color='grey', alpha=0.1, linestyle='--')
axes[1].axvline(x=2, color='grey', alpha=0.1, linestyle='--')
axes[1].set_xlim(-0.005, 2.005)
axes[1].set_ylim(bottom=-2.05, top=2.05)

plt.show()
```

where 

\begin{flalign}
\qquad A &= 2 & \\
\qquad f &= \frac{1}{2} \text{ Hz} &\\
\qquad \phi &= -\frac{\pi}{2} \text{ radians} & 
\end{flalign}

<hr style="width:50%;border-top:1px dashed;"/>

\begin{flalign}
\text{3.} \quad x(t) &= - \sin( 4\pi \cdot t + \pi) \qquad \rightarrow \qquad x(t) = \cos(2\pi \cdot 2 \cdot t - \frac{\pi}{2})&

#### Exercise 1.4

Given a sound source of constant power, and a microphone at distance $r$ observing sound pressure $p$:

\begin{flalign}
\text{1.} &\quad \text{since } p \propto \frac{1}{r} & \\
&\quad \frac{1}{2} \left( p \propto \frac{1}{r} \right) \quad \rightarrow \quad \frac{p}{2} \propto \frac{1}{2r} & \\
\end{flalign}

so moving the microphone to distance $2r$ should result in sound pressure $\frac{p}{2}$.

<hr style="width:50%;border-top:1px dashed;"/>

\begin{flalign}
\text{2.} &\quad \text{if sound pressure at } r \text{ is } L \text{[dB], the distance to produce } L - 5\text{[dB] can be found with:} & \\
& \quad 5 = 20 \cdot \log_{10}\frac{r + \alpha}{r}  & \\
& \quad 0.25 = \log_{10}\frac{r + \alpha}{r} & \\
& \quad \frac{r + \alpha}{r} \approx 1.778 \quad \rightarrow \quad \alpha \approx 0.778 \text{ } r&
\end{flalign}
b
So if having the microphone at distance $r$ from the sound source yields $L$ dB of sound pressure, moving the microphone an additional $0.778 \text{ }r$ away from the sound source should result in a decrease of $5$ dB.

#### Exercise 1.5

Given that the range of human hearing runs from 20 Hz to 20,000 Hz, approximately how many octaves does this span?

From [Donald Byrd's A Table of Musical Pitches](https://homes.luddy.indiana.edu/donbyrd/Teach/MusicalPitchesTable.htm), we assume that the lowest possible pitch for human hearing is $\text{E}_{0}$, at 20.602 Hz.

\begin{flalign}
\qquad \text{E}_{1} &= 20.602 \cdot 2^{1} = 41.204 & \\
\qquad \text{E}_{2} &= 20.602 \cdot 2^{2} = 82.408 & \\
\quad \vdots & \\
\qquad \text{E}_{9} &= 20.602 \cdot 2^{9} = 10548.224 & \\
\qquad \text{E}_{10} &= 20.602 \cdot 2^{10} = 21096.448 & \\
\end{flalign}

and so the highest audible pitch would be $\text{E}_{9}$, meaning that the range of human hearing is approximately 9 octaves.
